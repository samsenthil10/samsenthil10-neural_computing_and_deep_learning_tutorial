```python
import pandas as pd

ErrorMetrics = []

vgg16_accuracy = 0.6763
vgg16_precision = 0.6749
vgg16_recall = 0.6763
vgg16_f1 = 0.6753

vgg19_accuracy = 0.6491
vgg19_precision = 0.6499
vgg19_recall = 0.6491
vgg19_f1 = 0.6473

resnet50_accuracy = 0.4769
resnet50_precision = 0.479
resnet50_recall = 0.4769
resnet50_f1 = 0.469

inceptionv3_accuracy = 0.3541
inceptionv3_precision = 0.3395
inceptionv3_recall = 0.3541
inceptionv3_f1 = 0.3365

test_accuracy = [vgg16_accuracy, vgg19_accuracy, resnet50_accuracy, inceptionv3_accuracy]
test_precision = [vgg16_precision, vgg19_precision, resnet50_precision, inceptionv3_precision]
test_recall = [vgg16_recall, vgg19_recall,  resnet50_recall, inceptionv3_recall]
test_f1 = [vgg16_f1, vgg19_f1, resnet50_f1, inceptionv3_f1]

ErrorMetrics.append(test_accuracy)
ErrorMetrics.append(test_precision)
ErrorMetrics.append(test_recall)
ErrorMetrics.append(test_f1)

ErrorMetrics_pd = pd.DataFrame(ErrorMetrics)
ErrorMetrics_pd.columns = ["VGG16", "VGG19", "ResNet50", "InceptionV3"]
ErrorMetrics_pd = ErrorMetrics_pd.T
ErrorMetrics_pd.columns = ['Accuracy', 'Precision', 'Recall', 'F1 Score']

ErrorMetrics_pd
```

```python
import matplotlib.pyplot as plt
fig,ax = plt.subplots(figsize=(10,10))
ErrorMetrics_pd.plot.bar(width = 0.7, alpha = 0.7, cmap = 'coolwarm', ax = ax)
```
